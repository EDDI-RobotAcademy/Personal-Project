```python
from flask import Flask, render_template, request
app = Flask(__name__)

import time 
import serial

import threading

ser = serial.Serial('/dev/ttyACM0', 9600)
lock = threading.Lock()


def dataRecv():
    while ser.readable():
            data = ser.readline()
            msg = data.decode('utf-8')
            print(msg)
    return msg
            
read = threading.Thread(target = dataRecv)
read.start

            
@app.route('/')
def Fan_Project():
    return render_template('Fan_Project.html')

@app.route('/Manual_Mode',)
def Manual_Mode():
    lock.acquire()
    ser.write(b'1')
    lock.release()
    return render_template('Manual_Mode.html')

@app.route('/Auto_Mode')
def Auto_Mode():
    lock.acquire()
    ser.write(b'2')
    lock.release()
    return render_template('Auto_Mode1.html')

@app.route('/Servo_Start')
def manual_servo_start():
    lock.acquire()
    ser.write(b'3')
    lock.release()
    return render_template('Manual_Mode.html')
    
@app.route('/Servo_Stop')
def manual_servo_stop():
    lock.acquire()
    ser.write(b'4')
    lock.release()
    return render_template('Manual_Mode.html')

@app.route('/Bldc_Low')
def bldc_low():
    lock.acquire()
    ser.write(b'5')
    lock.release()
    return render_template('Manual_Mode.html')

@app.route('/Bldc_Midde')
def bldc_middle():
    lock.acquire()
    ser.write(b'6')
    lock.release()
    return render_template('Manual_Mode.html')

@app.route('/Bldc_High')
def bldc_high():
    lock.acquire()
    ser.write(b'7')
    lock.release()
    return render_template('Manual_Mode.html')

@app.route('/Bldc_Stop')
def bldc_stop():
    lock.acquire()
    ser.write(b'8')
    lock.release()
    return render_template('Manual_Mode.html')

@app.route('/Back')
def back():
    lock.acquire()
    ser.write(b'9')
    lock.release()
    return render_template('Fan_Project.html')


if __name__ == "__main__":
    app.run()
```
